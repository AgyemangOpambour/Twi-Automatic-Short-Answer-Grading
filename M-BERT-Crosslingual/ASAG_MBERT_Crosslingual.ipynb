```python
import numpy as np
import torch
from torch.utils.data import DataLoader, Dataset
from transformers import BertTokenizer, BertForSequenceClassification
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load and preprocess the Mohler dataset
df = pd.read_csv("/content/mohler_dataset_edited.csv")

# Step 2: Tokenize the input data
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

# Step 3: Define the dataset class
class GradingDataset(Dataset):
    def __init__(self, dataframe, tokenizer, max_length):
        self.data = dataframe
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        row = self.data.iloc[index]
        desire_ans = row['desired_answer']
        stud_ans = row['student_answer']
        score_avg = int(row['score_avg'])  # Convert label to integer

        encoding = self.tokenizer.encode_plus(
            desire_ans,
            stud_ans,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )

        input_ids = encoding['input_ids'].squeeze()
        attention_mask = encoding['attention_mask'].squeeze()

        return {
            'input_ids': input_ids,
            'attention_mask': attention_mask,
            'score_avg': score_avg
        }


# Step 4: Split the dataset into training, validation, and test sets

train_val_data, test_data = train_test_split(df, test_size=0.2, random_state=42)
train_data, val_data = train_test_split(train_val_data, test_size=0.1, random_state=42)

# Step 5: Create the datasets and data loaders
batch_size = 16
max_length = 128

train_dataset = GradingDataset(train_data, tokenizer, max_length)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

val_dataset = GradingDataset(val_data, tokenizer, max_length)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

test_dataset = GradingDataset(test_data, tokenizer, max_length)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Step 6: Create the model
num_labels = df['score_avg'].nunique()
model = BertForSequenceClassification.from_pretrained('bert-base-multilingual-cased', num_labels=num_labels)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model.to(device)

# Step 7: Define the optimizer and loss function
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-5)
loss_fn = torch.nn.CrossEntropyLoss()

# Step 8: Train the model
epochs = 6

for epoch in range(epochs):
    model.train()

    for batch in train_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['score_avg'].to(device)

        optimizer.zero_grad()

        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        logits = outputs.logits

        loss.backward()
        optimizer.step()

    model.eval()

    with torch.no_grad():
        val_loss = 0.0
        val_correct = 0

        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['score_avg'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            val_loss += outputs.loss.item()

            _, predicted = torch.max(outputs.logits, 1)
            val_correct += (predicted == labels).sum().item()

        val_loss /= len(val_loader.dataset)
        val_accuracy = val_correct / len(val_loader.dataset)

        print(f"Epoch {epoch+1}/{epochs}")
        print(f"Validation Loss: {val_loss:.4f}")
        print(f"Validation Accuracy: {val_accuracy:.4f}")

        # Compute predictions on the validation set
        predicted_labels = []
        true_labels = []

        for batch in val_loader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['score_avg'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
            _, predicted = torch.max(outputs.logits, 1)

            predicted_labels.extend(predicted.tolist())
            true_labels.extend(labels.tolist())

        # Compute and plot the confusion matrix
        cm = confusion_matrix(true_labels, predicted_labels)
        plt.figure(figsize=(10, 8))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.xlabel('Predicted Class')
        plt.ylabel('True Class')
        plt.title('Confusion Matrix')
        plt.savefig('/content/confusion_matrix.png')  # Save the plot as a PNG file
        plt.show()

        # Compute class-wise accuracy and plot the bar chart
        class_labels = sorted(df['score_avg'].unique())
        class_accuracy = []
        for class_label in class_labels:
            class_true = [true == class_label for true in true_labels]
            class_pred = [pred == class_label for pred in predicted_labels]
            class_acc = accuracy_score(class_true, class_pred)
            class_accuracy.append(class_acc)

        plt.figure(figsize=(10, 6))
        plt.bar(class_labels, class_accuracy)
        plt.xlabel('Class')
        plt.ylabel('Accuracy')
        plt.title('Class-wise Accuracy')
        plt.savefig('/content/classwise_accuracy.png')  # Save the plot as a PNG file
        plt.show()


```

```python
# Step 9: Save the model
torch.save(model.state_dict(), '/content/model.pth')

# Step 10: Export the test data as CSV
df_test = test_data[['question','desired_answer', 'student_answer', 'score_avg']]
df_test.to_csv('/content/test_data.csv', index=False)

# Step 11: Load the saved model and test data
model = BertForSequenceClassification.from_pretrained('bert-base-multilingual-cased', num_labels=num_labels)
model.load_state_dict(torch.load('/content/model.pth'))
model.to(device)

df_test = pd.read_csv('/content/trans_test_data_twi_en.csv')
```

```python
# Step 12: Preprocess and tokenize the test data
test_df = df_test.rename(columns={
    'Translated Question': 'question',
    'Translated Desired Answer': 'desired_answer',
    'Translated Student Answer': 'student_answer',
    'Average Score': 'score_avg'
})

# Step 12: Preprocess and tokenize the test data
test_dataset = GradingDataset(test_df, tokenizer, max_length)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Step 13: Perform inference on the test data
model.eval()
predicted_scores = []

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        _, predicted = torch.max(outputs.logits, dim=1)
        predicted_scores.extend(predicted.tolist())

# Step 14: Round up the predicted scores to two decimal places
rounded_scores = [round(score, 2) for score in predicted_scores]



# Step 15: Add predicted scores to the test data
df_test['Predicted Score'] = rounded_scores

# Step 16: Calculate the mean absolute error
from sklearn.metrics import mean_absolute_error

true_scores = df_test['Average Score'].values
predicted_scores = df_test['Predicted Score'].values

mae = mean_absolute_error(true_scores, predicted_scores)
print("Mean Absolute Error:", mae)

# Step 17: Drop the "score_avg" column
df_test.drop('Average Score', axis=1, inplace=True)

# Step 18: Save the predictions to a CSV file
df_test.to_csv('/content/pred_test_data_twi_en.csv', index=False)
```
